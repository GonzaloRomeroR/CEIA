```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

#### YPF dataset

```python
# Upload dataset
ypf_dataset = pd.read_csv("datasets/YPFD.2000.2021.csv")
# Transform fechaHora column to datetime
ypf_dataset["fechaHora"] = pd.to_datetime(ypf_dataset["fechaHora"])
# Drop zero values
ypf_dataset = ypf_dataset[ypf_dataset.ultimoPrecio != 0]
ypf_dataset.head()
```

```python
# Plot dataset
fig, ax = plt.subplots(figsize=(12, 12))

date_form = DateFormatter("%Y-%m-%d")
ax.xaxis.set_major_formatter(date_form)
plt.grid()
plt.plot(ypf_dataset["fechaHora"], ypf_dataset["ultimoPrecio"])
```

#### S1MME


```python
S1MME_dataset = pd.read_csv("datasets/S1MME_week44.csv")
S1MME_dataset.head()

# Upload dataset
S1MME_dataset = pd.read_csv("datasets/S1MME_week43.csv")
# Transform MyDay column to datetime
S1MME_dataset["MyDay"] = pd.to_datetime(S1MME_dataset["MyDay"])
S1MME_dataset.head()
```

```python

# Plot dataset
fig, ax = plt.subplots(figsize=(12, 12))

date_form = DateFormatter("%Y-%m-%d-%H")
ax.xaxis.set_major_formatter(date_form)
plt.grid()
for column in S1MME_dataset['NENAME'].unique():
    plt.legend(column)
    plt.plot(
        S1MME_dataset[S1MME_dataset['NENAME'] == column]["MyDay"], 
        S1MME_dataset[S1MME_dataset['NENAME'] == column]["S1_mode_combined_attach_request_times_SEQ"]
    )
```
